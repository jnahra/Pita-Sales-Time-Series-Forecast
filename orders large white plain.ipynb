```python
df_sales = pd.read_csv('large_white_plain_sales.csv')
```

```python
df_sales['date'] = pd.to_datetime(df_sales['date'])
df_sales.set_index('date',inplace=True)
```

```python
df_sales.head()
```

```python
df_sales['day'] = df_sales.index.dayofweek

df['day'] = df.index.dayofweek
```

```python
df.head(10)
```

```python
df_sales['quantity shifted'] = df_sales['Quantity'].shift(-1)
```

```python
df_sales['new quantity'] = df_sales['Quantity'] + df_sales['quantity shifted']
```

```python
df_sales.head(10)
```

```python
df_sales = df_sales.loc[(df_sales['day'] != 1) & (df_sales['day'] != 6)]
```

```python
def bake_days(df):
    if df['day'] == 5 or df['day'] == 0:
        return df['new quantity']
    else:
        return df['Quantity']        
```

```python
df_sales['sales'] = df_sales.apply(bake_days,axis=1)
```

```python
df_sales.head()
```

```python
df.head()
```

```python
df_so = pd.concat([df['orders'],df_sales['sales']],axis=1).dropna()
```
